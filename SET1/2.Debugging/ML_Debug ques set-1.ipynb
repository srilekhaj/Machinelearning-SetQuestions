```python
#10.
from sklearn.svm import SVC

X_train = [[1, 100], [2, 200], [3, 300]]
y_train = [0, 1, 0]

model = SVC()
model.fit(X_train, y_train)
print(model.predict([[1, 150]]))  # Unreliable output due to large-scale difference
```

```python
large difference can be minimized using the standardization technique
```

```python
#10.
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

X_train = [[1, 100], [2, 200], [3, 300]]
y_train = [0, 1, 0]

scalar = StandardScaler()
X_train_new = scalar.fit_transform(X_train)

model = SVC(kernel='linear', C=1)
model.fit(X_train_new, y_train)

input = [[1,150]]
scaled_input = scalar.transform(input)
print(scaled_input)
print(model.predict(scaled_input)) # Reliable output due to scaling
```
